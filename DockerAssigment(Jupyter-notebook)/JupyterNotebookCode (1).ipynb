```python
import pandas as pd
data = pd.read_csv("books.csv")
data.info()
```

```python
##print the first 5 rows
data.head()
```

```python
#change the desired data into numeric
data['ratings_count'] = pd.to_numeric(data['ratings_count'], errors='coerce')
data['average_rating'] = data['average_rating'].astype(float)

#clean these 3 columns from the null values they contain
data.dropna(subset=['title', 'ratings_count', 'average_rating'], inplace=True)

harry_potter_books = data[data['title'].str.contains("Harry Potter")]
most_selling_harry_potter = harry_potter_books.nlargest(10, 'ratings_count')


```

```python
# Find the most selling books within the Harry Potter series:
print("Most Selling Harry Potter Books:")
print(most_selling_harry_potter[['title', 'ratings_count']])
```

```python
#Calculate the average rating of the Harry Potter books.
average_rating_harry_potter = harry_potter_books['average_rating'].mean()
print("\nAverage Rating of Harry Potter Books:")
average_rating_harry_potter
```
